# Tiền xử lý
Có một số dữ liệu ta sẽ không cho vào trong model để train như Name vì nó không thể hiện được tính logic của bài toán ( dựa vào tên không thể biết được bạn chết hay sống)
Có một số dữ liệu như Sex, Embarked(điểm đến) không phải dạng số nên ta phải chuyển nó về dạng số để bỏ vào model

```python
# Chuyển dữ liệu
convert = pd.get_dummies(df.Embarked, prefix = 'Embarked')
convert
```

```python
# Xóa cột cũ
df.drop(["Embarked"], axis = 1, inplace = True)
```

```python
# Thêm vào dataset cột vừa xử lý
df = df.join(convert)
df
```

```python
df.Sex = df.Sex.map({'male': 0, 'female': 1})  # có thể dùng pd.get_dummies() như cột trên nhưng do chỉ có 2 giá trị nên ta set bằng pd.map() lun cho nhanh
df
```

```python
x = df.drop(['Survived'], axis = 1)
y = df.Survived.copy()
```

```python
# Những biến không thể hiện tính logic thì xóa đi
x.drop(['Cabin'],axis = 1,inplace = True)
x.drop(['Ticket'],axis = 1,inplace = True)
x.drop(['Name'],axis = 1,inplace = True)
x.drop(['PassengerId'],axis = 1,inplace = True)
```

```python
# Trong thực tế có một số chỗ mà chưa được điền dữ liệu ( missing value) , cách đơn giản nhất để xử lý việc này mà không làm mất số lượng dữ liệu là thay nó bằng giá trị tủng bình
x.isnull().values.any() # Check thử xem có missing value không
x[pd.isnull(x).any(axis = 1)] # Xem những hàng có missing value
```

```python
# Vậy là , ta tiến hành xử lý
x.Age.fillna(x.Age.mean(),inplace = True)
```

# Bắt đầu train

```python
from sklearn.linear_model import LogisticRegression # sử dụng pakkage hồi quy logictic
```

```python
# 80% train, 20% test
x_train, x_valid,y_train,y_valid = train_test_split(x,y,test_size = 0.2,random_state = 7)
```

```python
model = LogisticRegression() 
```

```python
model.fit(x_train,y_train) # fit model
```

```python
model.score(x_train,y_train ) # đánh gái mô hình
```

```python
b = model.intercept_ # bias
b = b[0]
b
```

```python
w = model.coef_ # hệ số weight[i]
w = w[0]
w
```

# Viết hàm dự đoán

#
Nhắc lại lý thuyết
Hàm dự đoán : f(z) = (1 + e ^ (- z)) ^ -1
    với z là hàm tuyến tính n biến
        Nếu f(z) >= 0.5 -> f(z) = 0 (True)
        else :             f(z) = 1 (False)

```python
def sizmoid(x):
    e = 2.71828182846
    sum_linear = b
    for i in range(len(w)):
        sum_linear+= w[i] * x[i]
    return 1 / (1 + e ** (- sum_linear ))
```

```python
def predict(x):
    if sizmoid(x) >= 0.5:
        return 0
    else:
        return 1
```
